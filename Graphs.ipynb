```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
rewards = pickle.load(open('./rewards/trained-model-750.pck', 'rb'))
n = 750
def get_avg_reward(last_avg):
    return [np.mean(rewards[i-last_avg:i]) for i in range(last_avg,n)]


fig, [(ax1, ax2), (ax3, ax4)] = plt.subplots(2, 2, figsize=(14, 6))
# fig.tight_layout()
plt.subplots_adjust(wspace = 0.5, hspace = 0.5)
ax1.plot(range(50, n), get_avg_reward(50))
ax1.set_title("Average Reward (50)")
ax1.set_ylabel("Avg Reward")
ax1.set_xlabel("Episodes")

ax2.plot(range(100, n), get_avg_reward(100))
ax2.set_title("Average Reward (100)")
ax2.set_ylabel("Avg Reward")
ax2.set_xlabel("Episodes")

ax3.plot(range(150, n), get_avg_reward(150))
ax3.set_title("Average Reward (150)")
ax3.set_ylabel("Avg Reward")
ax3.set_xlabel("Episodes")

ax4.plot(range(300, n), get_avg_reward(300))
ax4.set_title("Average Reward (300)")
ax4.set_ylabel("Avg Reward")
ax4.set_xlabel("Episodes")
```
